```python
import pandas as pd
from itertools import combinations
from collections import Counter
from typing import Tuple
from tqdm import tqdm
```

```python
df = pd.read_excel("/opt/ml/final-project-level3-recsys-07/crawler/codishop/codi_crawler/asset/codi_item_id.xlsx")
```

```python
len(df)
```

```python
dup = list()
gf = df.groupby(['id'])['item_id'].apply(list)
for val in tqdm(gf):
    val = sorted(val)
    comb_res = list(combinations(val, 2))
    dup.extend(comb_res)
```

```python
Counter(dup).most_common()
```
